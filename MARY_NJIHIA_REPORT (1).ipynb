 Overview

5.1 Interpretation of Model Performance

Three machine learning models were evaluated: Linear Regression, Decision Tree, and XGBoost.

The results showed noticeable differences in how each model handled the dataset:

- Linear Regression performed the weakest. This is expected because flood prediction is a non-linear and complex classification problem, while linear regression assumes a straight-line relationship between variables.

- Decision Tree performed better than linear regression because it can handle non-linear relationships and feature interactions. However, it showed signs of overfitting, meaning it learned the training data too well but did not generalize as effectively to unseen data.
- XGBoost delivered the best performance overall. It handled complex patterns in the dataset effectively by combining multiple weak learners into a strong ensemble model. It also reduced overfitting compared to a single decision tree, making it more reliable for flood prediction tasks.

Overall, the results confirm that ensemble methods are more suitable for flood prediction than simple linear.

5.4 Limitations of the Study

Although this study achieved meaningful results in flood prediction using machine learning models, several limitations were encountered that may have influenced the overall performance and generalizability of the findings.

# Limited Availability of High-Quality Data

One of the main limitations was the limited availability of comprehensive and high-resolution flood-related data. The dataset used did not include some important environmental variables such as:

Rainfall intensity and duration
River discharge levels
Soil moisture content
Elevation and drainage capacity

The absence of these features may have reduced the model’s ability to fully capture real flood dynamics.

# Data Imbalance and Distribution Issues

Flood datasets are often naturally imbalanced, with fewer flood occurrences compared to non-flood cases. This imbalance can lead to:

Bias toward the majority class
Reduced sensitivity in detecting rare flood events
Inconsistent model performance across different regions

Even though evaluation metrics like F1-score were used, imbalance still affected overall learning quality.

# Limited Impact of Population Data

Although population data was successfully merged with the flood dataset, its impact on improving predictive accuracy was moderate rather than significant. This is because population data mainly reflects exposure rather than direct flood causation, meaning it improves interpretability but not strongly the predictive signal.

# Model Constraints

The study relied on traditional machine learning models such as Linear Regression, Decision Tree, and XGBoost. While XGBoost performed best, limitations include:

Linear Regression is not suitable for complex non-linear flood patterns
Decision Trees are prone to overfitting
XGBoost, while powerful, still depends heavily on feature quality

More advanced models like deep learning (LSTM, CNNs) were not implemented due to computational and data constraints.

# Lack of Real-Time Data Integration

The model was trained using historical/static datasets and did not incorporate real-time environmental data such as:

Live rainfall measurements
Satellite imagery updates
River level monitoring systems

This limits its application in fully real-time flood forecasting systems.

# Geographic and Temporal Limitations

The dataset used may not fully represent all geographic regions or long-term climate variations. As a result:

Model performance may vary in different regions
Seasonal and long-term climate trends may not be fully captured
# Computational and Resource Constraints

Advanced techniques such as deep learning and large-scale data fusion were not applied due to limited computational resources and dataset size. This restricted exploration of more complex predictive architectures.


In summary, while the study provides a strong foundation for flood prediction using machine learning, its effectiveness is limited by data availability, model simplicity, and lack of real-time integration. Addressing these limitations in future work could significantly improve prediction accuracy and practical usability.
